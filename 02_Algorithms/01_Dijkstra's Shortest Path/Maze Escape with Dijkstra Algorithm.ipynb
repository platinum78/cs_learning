```python
from image_dijkstra import *
from matplotlib import pyplot as plt
```

```python
img_to_graph = ImageToGraph("./maze.png")
graph = img_to_graph.image_to_graph()
```

```python
plt.figure(1, figsize=[7, 7])
plt.imshow(img_to_graph.image)
plt.scatter(195, 190)
```

```python
dijkstra = DijkstraFinder(graph)
print("here!")
dijkstra.find_min_path(img_to_graph.idx_cvt(100, 100), img_to_graph.idx_cvt(190, 195))
```

```python
indices = []
for node in dijkstra.min_path:
    row = int(node.index / img_to_graph.width)
    col = node.index % img_to_graph.width
    indices.append((col, row))
```

```python
plt.figure(2, figsize=[10, 10])
plt.imshow(img_to_graph.image)
for index in indices:
    plt.scatter(index[0], index[1])
```
