## 6.2. Lists

### 6.2.1. Node-Based Operations and Iterators

* Containers and Positions
    * A list can be considered as a *container*.
        * Container is a gathering of any object.
    * A position is always defined in relative manner, in lists.
    * Position is not changed unless the related element is removed, since each node have their own linkage to next element.
    * If an element associated with a position is removed, the position is said to be *invalidated*.

* Iterators
    * Although position is useful, it would be even more useful if one might iterate through the container.
    * This idea gives the 'iterator'.
    * `begin()` and `end()` can serve as an initializer of iterator object.
