# 4.1. Analysis Tools

## 4.1. The Seven Functions Used in This Book

### 4.1.1. The Constant Function

For a fixed constant, the following notation is used.

$$ f(n) = c $$

Here, c is an arbitrary constant.

### 4.1.2. The Logarithm Function

There are many places that logarithmic functions appear.

$$ f(n) = \log_b {n} $$

### 4.1.3. The Linear Function

There are many cases that linear functions appear.

$$ f(n) = n $$

Here, the slope of the function is not important.

### 4.1.4. The N-Log-N Functions

$$ f(n) = n \log_b {n} $$

### 4.1.5. The Quadratic Function

$$ f(n) = n^2 $$

For nested loops, there are some cases that execution happens in n, n-1, n-2, ... 1. In this case, the total execution is given as follows.

$$ \frac{n^2 + n}{2} $$

This expression falls into the quadratic-function category.

### 4.1.6. Cubic and Other Polynomial Functions

* Cubic Functions

$$ f(n) = n^3 $$

* Summations

$$ f(n) = \sum_{i=0}^{d} {a_i n^i} $$

### 4.1.7. The Exponential Function

$$ f(n) = b^n $$

### 4.1.8. Comparison of Growth Rates

$$ c < \log {n} < n < n \log {n} < n^2 < n^3 < n^b $$

## 4.2. Analysis of Algorithms

* Data Structures: a systematic way of organizing and accessing data.
* Algorithms: step-by-step procedure for performing some task in a finite amount of time.
* To find out if data structures and algorithms are 'good', precise methods to assess them is essential.
